```python
x = [2023,2024,2025,2026]
y1 = [15,25,30,20]
y2 = [17,23,38,5]
y3 = [13,15,20,30]

plt.title("Class size",
          fontsize=25,
          family = "Arial",
          fontweight= "bold",
          color="#023e9e")

plt.xlabel("Year",fontsize=25,
                  family = "Arial",
                  fontweight= "bold",
                  color="#58a9bf")

plt.ylabel("Students",fontsize=25,
                  family = "Arial",
                  fontweight= "bold",
                  color="#58a9bf")

plt.tick_params(axis="both",
                colors="#013847")

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)

plt.xticks(x)

plt.subplots_adjust(left= 0.14,
                    bottom=0.175)

plt.show()
```
