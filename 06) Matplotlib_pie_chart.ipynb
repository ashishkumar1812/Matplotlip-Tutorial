```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
category = np.array(["Freshmen","Sophomores","Juniors","Seniors"])
values = np.array([300,250,275,225])
```

```python
plt.pie(values)
plt.show()
```

```python
plt.pie(values, labels=category)
plt.show()
```

```python
plt.pie(values, labels=category,
                autopct="%1f")
plt.show()
```

```python
plt.pie(values, labels=category,
                autopct="%1.1f")
plt.show()
```

```python
plt.pie(values, labels=category,
                autopct="%1.1f%%")
plt.show()
```

```python
category = np.array(["Freshmen","Sophomores","Juniors","Seniors"])
values = np.array([300,250,275,225])
colors = ["red","yellow","blue","green"]

plt.pie(values, labels=category,
                autopct="%1.1f%%",
                colors=colors)
plt.show()
```

```python
category = np.array(["Freshmen","Sophomores","Juniors","Seniors"])
values = np.array([300,250,275,225])
colors = ["red","yellow","blue","green"]

plt.pie(values, labels=category,
                autopct="%1.1f%%",
                colors=colors)
plt.title("Bro Code College")

plt.show()
```

```python
category = np.array(["Freshmen","Sophomores","Juniors","Seniors"])
values = np.array([300,250,275,225])
colors = ["red","yellow","blue","green"]

plt.pie(values, labels=category,
                autopct="%1.1f%%",
                colors=colors,
                explode= [0,0,0,0.1])

plt.title("Bro Code College")

plt.show()
```

```python
category = np.array(["Freshmen","Sophomores","Juniors","Seniors"])
values = np.array([300,250,275,225])
colors = ["red","yellow","blue","green"]

plt.pie(values, labels=category,
                autopct="%1.1f%%",
                colors=colors,
                explode= [0,0,0,0.1],
                shadow=True)

plt.title("Bro Code College")

plt.show()
```

```python
category = np.array(["Freshmen","Sophomores","Juniors","Seniors"])
values = np.array([300,250,275,225])
colors = ["red","yellow","blue","green"]

plt.pie(values, labels=category,
                autopct="%1.1f%%",
                colors=colors,
                explode= [0,0,0,0.1],
                shadow=True,
                startangle=90)

plt.title("Bro Code College")

plt.show()
```
