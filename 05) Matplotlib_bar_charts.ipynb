```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
category = np.array(["Grains","Fruit","Vegetable","Protien","Dairy","Sweets"])
values = np.array([4,3,2,5,3,1])
```

```python
plt.bar(category, values)
plt.show()
```

```python
plt.bar(category, values)

plt.title("Daily Consumption")
plt.xlabel("Food")
plt.show()
```

```python
plt.bar(category, values)

plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")
plt.show()
```

```python
plt.bar(category, values, color="skyblue")

plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")
plt.show()
```

```python
plt.barh(category, values, color="skyblue")

plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")
plt.show()
```
