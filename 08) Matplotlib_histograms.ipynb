```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
scores = np.random.normal(loc=80,scale=10,size=100)
```

```python
plt.hist(scores)
plt.show()
```

```python
scores = np.clip(scores,0,100)
plt.hist(scores)
plt.show()
```

```python
plt.hist(scores, bins=20)
plt.show()
```

```python
plt.hist(scores, bins=10)
plt.show()
```

```python
plt.hist(scores, bins=10,
                 color="lightgreen")
plt.show()
```

```python
plt.hist(scores, bins=10,
                 color="lightgreen",
                 edgecolor="black")
plt.show()
```

```python
plt.hist(scores, bins=10,
                 color="lightgreen",
                 edgecolor="black")
plt.title("Exam Scores")
plt.show()
```

```python
plt.hist(scores, bins=10,
                 color="lightgreen",
                 edgecolor="black")
plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("No. of students")
plt.show()
```
