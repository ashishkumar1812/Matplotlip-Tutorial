```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [5,10,15,20,25]

plt.grid()

plt.plot(x,y)
plt.show()
```

```python
plt.grid(axis="x")

plt.plot(x,y)
plt.show()
```

```python
plt.grid(axis="y")

plt.plot(x,y)
plt.show()
```

```python
plt.grid(axis="y",
         linewidth=2)

plt.plot(x,y)
plt.show()
```

```python
plt.grid(axis="y",
         linewidth=2,
         color="#abc722",)

plt.plot(x,y)
plt.show()
```

```python
plt.grid(axis="y",
         linewidth=2,
         color="#abc722",
         linestyle="dashed")

plt.plot(x,y)
plt.show()
```

```python
plt.grid(axis="y",
         linewidth=2,
         color="#abc722",
         linestyle="dotted")

plt.plot(x,y)
plt.show()
```

```python
plt.grid(axis="y",
         linewidth=2,
         color="#abc722",
         linestyle="dashdot")

plt.plot(x,y)
plt.show()
```
