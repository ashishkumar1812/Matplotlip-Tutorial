```python
x = [2023,2024,2025,2026]
y1 = [15,25,30,20]
y2 = [17,23,38,5]
line_style = dict(marker=".",
                  markersize = 30,
                  markerfacecolor = "cyan",
                  markeredgecolor = "cyan",
                  linestyle = "solid",
                  linewidth = 4,
                  color = "#0317fc")

plt.plot(x,y1,**line_style)
plt.plot(x,y2,**line_style)
plt.show()
```

```python
x = [2023,2024,2025,2026]
y1 = [15,25,30,20]
y2 = [17,23,38,5]
y3 = [13,15,20,30]
line_style = dict(marker=".",
                  markersize = 30,
                  markerfacecolor = "cyan",
                  markeredgecolor = "cyan",
                  linestyle = "solid",
                  linewidth = 4,
                  color = "#0317fc")

plt.plot(x,y1,**line_style)
plt.plot(x,y2,**line_style)
plt.plot(x,y3,**line_style)
plt.show()
```

```python
x = [2023,2024,2025,2026]
y1 = [15,25,30,20]
y2 = [17,23,38,5]
y3 = [13,15,20,30]
line_style = dict(marker=".",
                  markersize = 30,
                  markerfacecolor = "cyan",
                  markeredgecolor = "cyan",
                  linestyle = "solid",
                  linewidth = 4,)

plt.plot(x,y1,color = "#0317fc",**line_style)
plt.plot(x,y2,color = "#1aa60d",**line_style)
plt.plot(x,y3,color = "#fa0505",**line_style)
plt.show()
```
