```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df
```

```python
print(df["Type1"].value_counts())
```

```python
type_count = df["Type1"].value_counts()
```

```python
type_count = df["Type1"].value_counts()
plt.bar(type_count.index, type_count.values)
plt.show()
```

```python
type_count = df["Type1"].value_counts()
plt.barh(type_count.index, type_count.values)
plt.show()
```

```python
type_count = df["Type1"].value_counts(ascending=True)
plt.barh(type_count.index, type_count.values)
plt.show()
```

```python
type_count = df["Type1"].value_counts(ascending=True)
plt.barh(type_count.index, type_count.values, 
                           color="#03dffc", 
                           edgecolor="black")
plt.show()
```

```python
type_count = df["Type1"].value_counts(ascending=True)

plt.barh(type_count.index, type_count.values, 
                           color="#03dffc", 
                           edgecolor="black",)

plt.title("No. of pokemon by primary type")
plt.show()
```

```python
type_count = df["Type1"].value_counts(ascending=True)

plt.barh(type_count.index, type_count.values, 
                           color="#03dffc", 
                           edgecolor="black",)

plt.title("No. of pokemon by primary type")
plt.xlabel("Count")
plt.ylabel("Type")
plt.show()
```

```python
type_count = df["Type1"].value_counts(ascending=True)

plt.barh(type_count.index, type_count.values, 
                           color="#03dffc", 
                           edgecolor="black",)

plt.title("No. of pokemon by primary type")
plt.xlabel("Count")
plt.ylabel("Type")
plt.tight_layout()
plt.show()
```
